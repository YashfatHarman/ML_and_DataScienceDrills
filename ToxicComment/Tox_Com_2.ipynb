# 1. tokenize / Stop-word removal


```python
# a bare-bone tokenizer

def tokenizer(text):
    return text.split()

tokenizer("Runners like running and thus they run.")
```

# tokenizer_porter() and stop list will be used later in classifiers

```python
#porter-stemmer tokenizer

from nltk.stem.porter import PorterStemmer

porter = PorterStemmer()

def tokenizer_porter(text):
    return [porter.stem(word) for word in text.split()]


tokenizer_porter("Runners like running and thus they run.")
```

```python
#function to stop-word removal

import nltk
#nltk.download("stopwords")

from nltk.corpus import stopwords
stop = stopwords.words("english")

[w for w in tokenizer_porter("a runner likes running and runs a lot") if w not in stop ]
```

# train a tfidf-vectorizer first and then use it for the rest of the classifiers

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#just for test, lets only select the X values for which y is labeled as toxic. Lets see how tfidf performs for those.
#X_toxic = X[y_toxic==1]
#print(len(X_toxic))
#X_toxic[:10]
```

```python
import sys
sys.setrecursionlimit(3000) # 3000 was proved to be enough to apply tfidf on the entire X_train in VBox. Need to check if th behavior changes in Tesla.

tfidf = TfidfVectorizer(strip_accents = None, lowercase = False, preprocessor = None, ngram_range = (1,1), 
                        stop_words = stop, tokenizer = tokenizer_porter, use_idf = True)

X_train_tfidf = tfidf.fit_transform(X_train)
#got a max recursion limit reached with the complete dataset. Will check later.
#probably just have to increase stack length.
```

```python
#now get tfidf for the test data as well
sys.setrecursionlimit(5000) #3000 didn't work for test. So tried 4000 and eventually 5000 worked. 
X_test_tfidf = tfidf.transform(X_test)
```

# Got tfidf. now do regression with it.

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

from sklearn.model_selection import cross_val_score
```

```python
lr = LogisticRegression(random_state = 0, penalty = "l2", C = 10.0)
```

```python
#scores = cross_val_score(estimator = lr, X = X_train_tfidf, y = y_toxic_train, cv = 10, scoring = "accuracy", n_jobs = -1)
lr = lr.fit(X_train_tfidf, y_toxic_train)
```

```python
lr_score_train = lr.score(X_train_tfidf, y_toxic_train)
lr_score_test = lr.score(X_test_tfidf, y_toxic_test)
print("Logistic regression: ")
print("Accuracy score, train: {:.2f}".format(lr_score_train))
print("Accuracy score, test: {:.2f}".format(lr_score_test))
```

# perceptron

```python
from sklearn.linear_model import Perceptron

pt = Perceptron(random_state = 0, class_weight = "balanced")
pt = pt.fit(X_train_tfidf, y_toxic_train)

pt_score_train = pt.score(X_train_tfidf, y_toxic_train)
pt_score_test = pt.score(X_test_tfidf, y_toxic_test)
print("Perceptron: ")
print("Accuracy score, train: {:.2f}".format(pt_score_train))
print("Accuracy score, test: {:.2f}".format(pt_score_test))
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn = knn.fit(X_train_tfidf, y_toxic_train)

knn_score_train = knn.score(X_train_tfidf, y_toxic_train)
knn_score_test = pt.score(X_test_tfidf, y_toxic_test)
print("KNN: ")
print("Accuracy score, train: {:.2f}".format(knn_score_train))
print("Accuracy score, test: {:.2f}".format(knn_score_test))
```
